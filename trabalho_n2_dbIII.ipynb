```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/ntcdave/trabalho-n2-dbII/main/Athens%202004%20Olympics%20Nations%20Medals.csv')


def generate_medal_pie_chart(df):
    # Calculamos o total de medalhas por pas
    medal_counts = df.set_index('NOC')['Total'].sort_values(ascending=False)

    # Pegamos os top 10 pases para manter o grfico legvel
    top_10 = medal_counts.head(10)

    # Somamos o resto em "Outros"
    others = pd.Series({'Outros': medal_counts.iloc[10:].sum()})

    # Combinamos os top 10 com "Outros" usando concat
    pie_data = pd.concat([top_10, others])

    # Calculamos as porcentagens
    percentages = pie_data / pie_data.sum() * 100

    # Criamos o grfico de pizza
    plt.figure(figsize=(12, 8))
    plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=90)
    plt.title('Distribuio de Medalhas por Pas (Top 10 + Outros)')
    plt.axis('equal')
    plt.show()

# Chamada da funo
generate_medal_pie_chart(df)
```

O grfico gerado pela funo generate_medal_pie_chart apresenta a distribuio percentual do total de medalhas conquistadas pelos 10 pases que mais ganharam medalhas, somando todas as medalhas dos demais pases na categoria "Outros". Assim, o grfico de pizza facilita a visualizao comparativa da quantidade de medalhas obtidas pelos principais pases em relao ao total de medalhas distribudas, proporcionando uma compreenso clara de quais naes dominaram na conquista de medalhas.
