```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/ntcdave/trabalho-n2-dbII/main/Athens%202004%20Olympics%20Nations%20Medals.csv')


def generate_medal_pie_chart(df):
    # Calculamos o total de medalhas por país
    medal_counts = df.set_index('NOC')['Total'].sort_values(ascending=False)

    # Pegamos os top 10 países para manter o gráfico legível
    top_10 = medal_counts.head(10)

    # Somamos o resto em "Outros"
    others = pd.Series({'Outros': medal_counts.iloc[10:].sum()})

    # Combinamos os top 10 com "Outros" usando concat
    pie_data = pd.concat([top_10, others])

    # Calculamos as porcentagens
    percentages = pie_data / pie_data.sum() * 100

    # Criamos o gráfico de pizza
    plt.figure(figsize=(12, 8))
    plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=90)
    plt.title('Distribuição de Medalhas por País (Top 10 + Outros)')
    plt.axis('equal')
    plt.show()

# Chamada da função
generate_medal_pie_chart(df)
```

O gráfico gerado pela função generate_medal_pie_chart apresenta a distribuição percentual do total de medalhas conquistadas pelos 10 países que mais ganharam medalhas, somando todas as medalhas dos demais países na categoria "Outros". Assim, o gráfico de pizza facilita a visualização comparativa da quantidade de medalhas obtidas pelos principais países em relação ao total de medalhas distribuídas, proporcionando uma compreensão clara de quais nações dominaram na conquista de medalhas.


```python
# @title Medalhas de Ouro

from matplotlib import pyplot as plt
df['Gold'].plot(kind='hist', bins=20, title='Medalhas de Ouro')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

Este gráfico mostra a distribuição das medalhas de ouro. O gráfico utiliza a coluna 'Gold' do DataFrame para criar 20 intervalos (bins) que agrupam as contagens de medalhas. O título do gráfico é simplesmente "Medalhas de Ouro". A remoção das bordas superior e direita (spines) dá ao gráfico um aspecto mais limpo e moderno, focando a atenção nos dados. Este histograma permite visualizar rapidamente a frequência de diferentes quantidades de medalhas de ouro, mostrando se há uma distribuição uniforme ou se existem picos em certos intervalos.
